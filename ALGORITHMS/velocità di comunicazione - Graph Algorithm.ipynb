```python
from array import array
from collections import deque

def velocita_di_comunicazione(lista_adiacenza, nodo_partenza, nodo_arrivo):
    
    
    if nodo_partenza == nodo_arrivo: 
        return 0
    
    numero_nodi = len(lista_adiacenza)
    nodi_visitati = array('b', [False] * numero_nodi)                         # array dei nodi visitati  O(n)
    lunghezza_max = float("inf")                                              # lunghezza del percorso più breve inizializzata a infinito
    capacita_minima = float('inf')                                            # capacità minima degli archi inizializzata a infinito
     
    coda_percorsi = deque()                                                            # coda per i possibili percorsi tra un nodo e l'altro
    coda_percorsi.append(array('I', [nodo_partenza]))  
   
 
    while coda_percorsi:                                                               # finchè la coda non si svuota      O(p)
        
        lista_percorso = coda_percorsi.popleft()      # estraggo dalla coda il primo elemento. ogni elemento nella coda è formato da: ultimo nodo del percorso, lunghezza del percorso, capacità minima del percorso
        nodo_corrente = lista_percorso[0]         # il nodo corrente sarà il secondo elemento dell'array
        
        if len(lista_percorso) >  1:                  # se la lista contiene più di un elemento, quindi non c'è solo il nodo iniziale
            lunghezza_percorso = lista_percorso[1]   # la lunghezza del percorso sarà il secondo elemento dell'array
                                                        
            
        else:
            lunghezza_percorso = 1
            
        
        nodi_visitati[nodo_corrente] = True                  # marchiamo quel nodo come visitato, per evitare di rivisitarlo e creare loop

        lunghezza_lista_adiacenza_nodo_corrente = len(lista_adiacenza[nodo_corrente])   #impostiamo la lunghezza della lista di adiacenaza del nodo corrente  O(n)                                 
        lista_adiacenza_nodo_corrente = lista_adiacenza[nodo_corrente]               # estraiamo l'array corrispondenente alla lista di adiacenza di un nodo per evitare di rifarlo più volte successivamente    O(n)                     
            
        for i in range(0, lunghezza_lista_adiacenza_nodo_corrente, 2):    # per tutti i nodi adiacenti al nodo corrente    O(m)
            nodo_adiacente = lista_adiacenza_nodo_corrente[i]        # mi prendo il nodo adiacente e il peso dell'arco che lo collega al nodo_corrente
            capacita_arco = lista_adiacenza_nodo_corrente[i + 1]
        

            if not nodi_visitati[nodo_adiacente]:         # proseguo ad analizzare questo nodo adiacente solo se non l'ho già visitato
                
                if len(lista_percorso) == 1:                  # se la lunghezza della lista_percorso è uguale a 1, quindi stiamo visitando i nodi adiacenti del nodo di partenza
                    capacità_min_percorso = capacita_arco   # la capacità minima del percorso è uguale alla capacità dell'arco verso il suo vicino in considerazione
                   
                else:           
                    capacità_min_percorso = lista_percorso[2]     #altrimenti prendo la capacità minima dalla lista del percorso, (il terzo elemento dell'array) e la confronto con il peso dell'arco in considerazione
                    if capacita_arco < capacità_min_percorso:      # se questo arco ha peso minore aggiorno la capacità minima del percorso
                        capacità_min_percorso = capacita_arco
            
                
                if nodo_adiacente == nodo_arrivo:                    # se arriviamo al nodo di arrivo
                    lunghezza_max = lunghezza_percorso+1                   # imposto la lunghezza massima dei percorsi che congiungono partenza e arrivo, per limitare la ricerca dei percorsi e non sprecare risorse
                    if capacità_min_percorso < capacita_minima:       # se la capacità minima del percorso prima trovata è minore della capacità minima generale, aggiorno quest'ultima 
                        capacita_minima = capacità_min_percorso
                        
                    
                elif nodo_adiacente != nodo_arrivo and (lunghezza_percorso+1) <= lunghezza_max:   #se non sono ancora arrivato al nodo finale e se la lunghezza del nuovo percorso che sto per creare rientra nella lunghezza massima dei percorsi
                    
                    nuova_lunghezza = lunghezza_percorso+1      # la lunghezza del percorso che sto per creare
                    
                    coda_percorsi.append(array("I", [nodo_adiacente, nuova_lunghezza, capacità_min_percorso]))                               # metto il percorso con l'aggiunta alla fine della coda dei percorsi
                    
                    
    return capacita_minima

            
            
    




lista_adiacenza =deque([
    array("I", []),                                   # Nodo 0
    array("I", [2, 1, 5, 3]),                         # Nodo 1
    array("I", [1, 1, 3, 8, 6, 7]),                   # Nodo 2
    array("I", [2, 8, 4, 3, 7, 2]),                   # Nodo 3
    array("I", [3, 3, 8, 9]),                         # Nodo 4
    array("I", [1, 3, 6, 4, 9, 6]),                   # Nodo 5
    array("I", [2, 7, 5, 4, 7, 5, 10, 8]),            # Nodo 6
    array("I", [3, 2, 6, 5, 8, 6, 11, 5]),            # Nodo 7
    array("I", [4, 9, 7, 6, 12, 2]),                  # Nodo 8
    array("I", [5, 6, 10, 1]),                        # Nodo 9
    array("I", [6, 8, 9, 1, 11, 2]),                  # Nodo 10
    array("I", [7, 5, 10, 2, 12, 5]),                 # Nodo 11
    array("I", [8, 2, 11, 5])                         # Nodo 12
])


velocita_di_comunicazione(lista_adiacenza,10,5)



```

Sapendo che la società Fraudolent garantisce una trasmissione che passa per un percorso che unisce due utenti con il minor numero di nodi intermedi, ma la cui velocità massima corrisponde alla capacità minima degli archi di quel percroso, per scoprire a quale velocità navigheranno due utenti (corrispondenti a due nodi distinti del grafico) possiamo implementare il seguente algoritmo:

Innanzitutto forniamo in input alla funzione "velocita_di_comunicazione()" la lista di adiacenza del nostro grafo, inizializzata come una coda di array, nella quale (a partire dal nodo 0 anche se non esiste, per semplificare il codice) inseriamo i nodi adiacenti ad ogni altro nodo, seguiti dal peso dell'arco che li unisce. Per esempio la lista che contiene i vertici del nodo 6 overo "array("I", [2, 7, 5, 4, 7, 5, 10, 8])" ci indica che il nodo 6 è adiacente ai nodi 2,5,7,10, e il peso degli archi tra il nodo 6 e il corrispoettivo nodo adiacente è 7,4,5,8.

In input avrei potuto fornirgli anche una matrice di adiacenza ma il costo temporale e spaziale sarebbero aumentati, mentre la lista di adiacenza in questo caso è più conveniente grazie proprio all'efficienza delle sue operazioni quando c'è bisogno di visitare un grafo e trovare gli archi incidenti sui nodi.

Come secondo e terzo input forniamo alla funzione i numeri del nodo iniziale e di quello finale.

Il comportamento dell'algoritmo è il seguente: viene eseguita una visita in ampiezza (BFS) del grafo, usando una coda per memorizzare di volta in volta i percorsi costruiti.
finché la coda non si svuota, estraiamo ed eliminiamo dalla coda il primo elemento, ovvero un percorso, e prediamo l'ultimo nodo. A questo punto, se siamo arrivati al nodo finale, ci calcoliamo la capacità minima in quel percorso, altrimenti proseguiamo nella BFS espandendo i nostri percorsi dai nodi adiacenti all'ultimo nodo di ogni percorso, appunto in ampiezza (controllando sempre di non aver già visitato il nodo, altrimenti rischiamo di creare loop infini. 

Tornando alla capacità minima, la possiamo trovare calcolandoci per ogni nodo del percorso in considerazione, il peso degli archi per poi prendere quello minore dopo averlo confrontato con la variabile di controllo "capacita_minima". 
In questo modo, all'interno di ogni percorso con il numero di nodi minimo che collega i due utenti, trovo la capacità minima di quello stesso percorso e infine la confronto con le capacità minime degli altri percorsi più brevi. La più piccola di tutte corrisponderà proprio alla velocità di navigazione offerta dalla Fraudolent tra i due utenti. 

Per quanto riguarda il Costo Temporale, le prime righe di codice hanno costo costante O(1), come anche il resto delle operazioni eseguite sulla coda, le restanti righe si trovano dentro a un ciclo while. Quest'ultimo viene eseguito finché non si svuota la coda, e quindi viene eseguito P volte, dove con P indichiamo il numero massimo di percorsi possibili

Al suo interno, nel caso in cui non siamo ancora arrivati al nodo finale, creaiamo un nuovo percorso e lo inseriamo nella coda. Questa operazione Richiede un tempo proporzionale alla lunghezza del percorso corrente, che indichiamo con L.

Invece, nel caso in cui arriviamo al nodo finale, viene eseguito un ciclo for su tutti i nodi del percorso per calcolare la capacità degli archi del percorso, quindi con costo nel caso peggiore sempre pari a O(L). Dopodichè viene effettuato un altro ciclo for per trovae i nodi adiacenti al nodo in considerazione, il quale nel caso peggiore ha costo O(m). Quindi il costo dei cicli for dentro il while nel caso in cui arrviamo al nodo finale, nel caso peggiore è O(L * m)
Il costo totale sarà quindi dato dalla ripetizione del costo precedente P volte, ovvero per ogni percorso che aggiungiamo nella coda dei percorsi, quindi O(P * L * m)

Se vogliamo esprimerlo solamente in termini di nodi e archi, dobbiamo tener conto che il numero massimo di percorsi è dato da n^2 e la lunghezza massima di un percorso è data dal numero degli archi, quindi da m. Il costo finale in termini di nodi e archi è quindi: O(n^2 * m^2) 

Un costo temporale così elevato è dovuto al fatto che dobbiamo trovare tutti i percorsi più brevi che connettono i due utenti e per ognuno di essi dobbiamo trovare il peso degli archi che li compongono e selezionare solo il più piccolo. 

Per quanto riguarda il Costo Spaziale, dobbiamo tenere in considerazione principalmente l'input che gli forniamo, ovvero la lista di adiaceanza, nel quale avremo n righe, quindi O(n), una per ogni nodo, e m archi, distribuiti nelle righe (il numero di nodi adiacenti per ogni riga corrisponde al grado di ogni nodo) e che rientrano nell'ordine di O(m). Quindi il costo spaziale della lista di adiacenza è O(n+m), il che è conveniente rispetto all'utilizzo di una matrice di adiacenza/ incidenza che avrebbe avuto costo spaziale O(n^2) / O(nm). Infatti se utilizziamo una matrice dobbiamo segnare anche gli 0 nel caso in cui due nodi/ un arco e un nodo non siano adiacanti/incidenti. Invece con una lista di adiacenaza dobbiamo solo memorizzare i nodi adiacenti (anche aggiungendo il peso degli archi, il costo spaziale sarebbe 2m che rientra comunque nell'ordine di O(m) ). 
Le variabili di controllo hanno costo spaziale costante, l'array dei nodi visitati ha costo O(n). La dimensione massima della coda dipende dalla lunghezza del percorso più breve tra i nodi di partenza e arrivo, quindi nel caso peggiore in cui i percorsi più brevi includono tutti i nodi nel grafo, il costo spaziale è O(n). il resto delle variabili, anche se variano in contenuto durante l'esecuzione dell'algoritmo, hanno costo costante. 

Il costo Spaziale complessivo finale nel caso peggiore è quindi O(n+m).
